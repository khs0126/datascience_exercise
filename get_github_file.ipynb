## 어떻게 하면 될까? 
- URL 을 따라 들어간 github 웹페이지를 보면 우상단쯤에 raw라는 버튼이 있다.
- raw 버튼을 클릭하고 주소창의 URL 을 가져오자

    https://raw.githubusercontent.com/keras-team/keras/master/examples/mnist_cnn.py

## 이 URL 을 통해 다운로드 받고 keras 를 실행해보자
